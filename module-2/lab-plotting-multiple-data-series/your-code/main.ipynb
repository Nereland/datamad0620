```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
#Esto es porque te gustó el otro día:
with plt.xkcd():
    toplot = data.groupby(by= "ItemType").agg({"RetailSales": "count", "RetailTransfers":"count","WarehouseSales":"count"})
    toplot.plot.bar()
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
data.columns
```

```python
data["sum_sales"]=(data['RetailSales']+data['RetailTransfers']+data['WarehouseSales']).sort_values(ascending=False)

data = data.sort_values(by='sum_sales', ascending=False)
bestsellers = data[:10]


toplot = bestsellers["sum_sales"].plot.barh(y="sum_sales")
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
average = data.groupby(by= "Month").agg({"RetailSales": "mean", "RetailTransfers":"mean","WarehouseSales":"mean"})
average.plot()
```

## 4. Plot the same information as above but as a bar chart.

```python
average.plot.bar()
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
import seaborn as sns; sns.set(color_codes=True)

sns.lmplot(x="RetailSales", y="RetailTransfers", hue = "ItemType", data = data)


```
